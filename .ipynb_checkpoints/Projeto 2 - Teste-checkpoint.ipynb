### Regressão Logística

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
```

```python
var_target = dados.output
var_logist = dados.drop(['output'], axis=1)
```

```python
x_treino, x_teste, y_treino, y_teste = train_test_split(var_logist, var_target, test_size = 0.3)
```

```python
logistic = LogisticRegression()
logistic.fit(x_treino, y_treino)
predicao = logistic.predict(x_teste)
predicao
```

```python
print(f'Acuracia da Regressão Logística com o treino: {(logistic.score(x_treino, y_treino)*100).round(2)}%.')
print(f'Acuracia da Regressão Logística com o teste: {(logistic.score(x_teste, y_teste)*100).round(2)}%.')
```

```python
confusion_matrix(y_teste, predicao)
```

```python
import statsmodels.api as sm
logit_model=sm.Logit(y_treino,x_treino)
result=logit_model.fit(method='bfgs')
result.summary2()
```

### Árvore de decisão 

```python
from sklearn.tree import DecisionTreeClassifier 
clf = DecisionTreeClassifier()

# Classificador Árvore de Decisão
clf = clf.fit(x_treino,y_treino)

pred1 = clf.predict(x_teste)
```

```python
from sklearn import metrics 
print("Acuracia:",((metrics.accuracy_score(y_teste, pred1)*100).round(2)))
```

```python
confusion_matrix3=confusion_matrix(y_teste,pred1)
print(confusion_matrix3)
print ('Predições corretas: {}'.format(confusion_matrix3[0][0]+confusion_matrix3[1][1]))
print ('Predições incorretas: {}'.format(confusion_matrix3[0][1]+confusion_matrix3[1][0]))
```

```python
from sklearn import tree
fig = plt.figure(figsize=(50,40))
_ = tree.plot_tree(clf, 
                   feature_names=x_treino.columns,  
                   class_names=['0','1'],
                   filled=True)
```
